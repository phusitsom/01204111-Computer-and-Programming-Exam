## 10. What is the output of the following program if suppose that hte user enters **3.2** as an input?

````python
def compute_circle(r):
    area = 3.14*r**2
    return area
r = int(input('Enter a radius: '))
area = compute_circle(r)
print(f'Area of a circle with {r} is {area:.2f}')
````

## 58. Let **n**  be equal to **3**, which pair of the following if-else blacks give the same print out for whatever value of **n**?

### Code

````python
def A(n):
    if n==1 and n==2:
        print('ant')
    elif n==3 and n==4 and n==5:
        print('bee')
    elif n==6 and n==7:
        print('butterfly')
    else:
        print('cockroach')
        
def B(n):
    if n==1 or n==2:
        print('ant')
    elif n==3 or n==4 or n==5:
        print('bee')
    elif n==6 or n==7:
        print('butterfly')
    else:
        print('cockroach')
        
def C(n):
    if 1<=n or n<=2:
        print('ant')
    elif 3<=n or n<=5:
        print('bee')
    elif 6<=n or n<=7:
        print('butterfly')
    else:
        print('cockroach')
        
def D(n):
    if 1<=n and n<=2:
        print('ant')    
    elif 3<=n and n<=5:
        print('bee')
    elif 6<=n and n<=7:
        print('butterfly')
    else:
        print('cockroach')
````

### Test

````python
for n in [1,3,5,7,9]:
    A(n)
    B(n)
    C(n)
    D(n)
    print()
````
